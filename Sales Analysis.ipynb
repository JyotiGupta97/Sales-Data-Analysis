<span style="font-size:20px; font-weight:bold;">Gender</span>


```python
ax = sns.countplot(x = 'Gender', data = df)
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)

ax.set_ylabel('Count of Orders Placed')
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x= 'Gender', y= 'Amount', data = sales_gen)
```

From the above 2 graphs it is observed that women shop more than men and they spend more money for shopping as well.

<span style="font-size:20px; font-weight:bold;">Age</span>


```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)

ax.set_ylabel('Count of Orders Placed')
```

This graph shows that the maximum amount of orders are placed by the age group 26-35 in which women place more number of orders than men.

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x= 'Age Group', y= 'Amount', data = sales_age)
```

From the above graph, age 26-35 has spent the most amount of money on shopping.

<span style="font-size:20px; font-weight:bold;">State</span>

```python
#Total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize': (17,5)})
sns.barplot(data = sales_state, x= 'State', y= 'Orders')
```

```python
#Total amount/sales from 10 states
sales_state = df.groupby(['State'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data = sales_state, x = 'State',y='Amount')
```

From the 1st graph we can see the top 10 states where the orders are being placed from in descending order. 
From the second graph we can see the top 10 states and the amount being spent on orders.
However, in the first graph the 3rd state from the right is Kerala and in the second graph the 3rd state from the right is Haryana. This tells us that despite the number of orders placed from Kerala are higher than Haryana, the amount spent on orders is higher in Haryana than in Kerala.

<span style="font-size:20px; font-weight:bold;">Marital Status</span>

```python
ax = sns.countplot(x = 'Marital_Status', data = df)
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)

ax.set_ylabel('Count of Orders Placed')
ax.set_xticklabels(['Married', 'Unmarried'])
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize': (7,5)})
sns.barplot(data = sales_state, x= 'Marital_Status', y= 'Amount', hue='Gender')
```

From the above graphs we can see that most of the buyers are married and of those, most are women and they have high purchasing power.

<span style="font-size:20px; font-weight:bold;">Occupation</span>

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
        ax.bar_label(bars)
ax.set_ylabel('Count of Orders Placed')
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize': (20,5)})
sns.barplot(data = sales_state, x= 'Occupation', y= 'Amount')
```

From the above graphs we can see that most of the buyers are from IT, Healthcare, and Aviation sector.

<span style="font-size:20px; font-weight:bold;">Product Category</span>

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

From above graphs, we see that most of the products sold are in the category Food, Clothing and Electronics.

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

<span style="font-size:20px; font-weight:bold;">Conclusion</span>

*Married women of age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category.*
